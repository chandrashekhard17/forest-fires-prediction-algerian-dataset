# Exploratory data analysis


```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

# missing values


```python
dataset[dataset.isnull().any(axis=1)]
```

Divide this dataset into the 2 regions from the column number is 122

```python
dataset.loc[:122,'Region']=0
dataset.loc[122:,'Region']=1
```

```python
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

```python
df.head()
```

```python
df[['month', 'day', 'year', 'Temperature','RH', 'Ws']] = df[['month', 'day', 'year', 'Temperature','RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
objects=[features for features in df.columns if df[features].dtypes=='O'] 
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
objects
```

```python
df.describe()
```

# Monthly Fire Analysis

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
dftemp=df_copy.loc[df_copy['Region']==0]
plt.subplots(figsize=(12,7))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.xlabel('Number of Fires happening in the Months',weight='bold')
plt.ylabel('Number of counting on the forest fires',weight='bold')
plt.title('Fire analysis for Beijia Region',fontsize=18,weight='bold')
plt.xticks(np.arange(4),['June','July','August','September'])
plt.grid(alpha=0.5)
plt.show()
```

```python
dftemp=df_copy.loc[df_copy['Region']==1]
plt.subplots(figsize=(12,7))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.xlabel('Number of Fires happening in the Months',weight='bold')
plt.ylabel('Number of counting on the forest fires',weight='bold')
plt.title('Fire analysis for Sidi Belabbes region',fontsize=18,weight='bold')
plt.xticks(np.arange(4),['June','July','August','September'])
plt.grid(alpha=0.5)
plt.show()
```

* Its observed that August and September had the most number of forest fires for both regions. And from the above plot of months, we can understand few things

* Most of the fires happened in August and very high Fires happened in only 3 months - **June, July and August.**

* Less Fires was on September
